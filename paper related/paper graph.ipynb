```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load datasheet
df = pd.read_csv("Book2.csv")

wavelength = df["Lamda"]
qe = df["QE"]
energy = df["Photon Energy"]

plt.figure(figsize=(10,6))

# Scatter plot with photon energy color scale
sc = plt.scatter(
    wavelength,
    qe,
    c=energy,
    cmap="viridis",
    s=35,
    alpha=0.8
)

# Colorbar
cbar = plt.colorbar(sc)
cbar.set_label("Photon Energy (eV)")

# Axis labels
plt.xlabel("Wavelength Spectrum (nm)")
plt.ylabel("Quantum Efficiency (%) / QE")

# Grid and limits
plt.grid(True, linestyle="--", alpha=0.5)
plt.xlim(290, 820)
plt.ylim(50, 120)

# Key points shown in the figure
points = [
    (300, 89.64, 4.13),
    (400, 99.96, 3.10),
    (500, 99.88, 2.48),
    (600, 99.65, 2.07),
    (700, 98.95, 1.77),
    (800, 96.58, 1.55)
]

for x, y, e in points:
    plt.scatter(x, y, color="black", s=60, zorder=5)
    plt.text(x, y+3, f"{y}\n{e}eV", ha="center", fontsize=9)

# Title
plt.title("Spectral Quantum Efficiency Analysis")

plt.tight_layout()
plt.show()
```
